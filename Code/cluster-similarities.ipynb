```python
import pandas as pd
import numpy as np
import pickle
from itertools import combinations
from sklearn.metrics import jaccard_score
```

```python
email_cluster = pd.read_pickle('Email/email_cluster_bert.pkl')
email_cluster
```

```python
podcast_cluster = pd.read_pickle('Podcast/podcast-bert-clust.pkl')
podcast_cluster
```

```python
tv_clusters = pd.read_pickle('/scratch/rm1667/rm1667-outputs/TV/tv-bert-clust.pkl')
tv_clusters
```

```python
email_words = email_cluster[['Name', 'Representation']].T.drop(columns=[0]).T
```

```python
podcast_words = podcast_cluster[['Name', 'Representation']].T.drop(columns=[0]).T
```

```python
tv_words = tv_clusters[['Name', 'Representation']].T.drop(columns=[0]).T
```

```python
import pandas as pd
from itertools import product

def calculate_jaccard_similarity(email_words, podcast_words, tv_words):
    result = []
    
    # Internal Comparison within Email Words DataFrame
    for idx1, idx2 in product(email_words.index, repeat=2):
        if idx1 != idx2:
            topic1_words = set(email_words.loc[idx1, 'Representation'])
            topic2_words = set(email_words.loc[idx2, 'Representation'])
            jaccard_score = len(topic1_words.intersection(topic2_words)) / len(topic1_words.union(topic2_words))
            result.append({'Source1': 'Email', 'Topic1': idx1-1, 'Source2': 'Email', 'Topic2': idx2-1, 'Similarity': jaccard_score})

    # Internal Comparison within Podcast Words DataFrame
    for idx1, idx2 in product(podcast_words.index, repeat=2):
        if idx1 != idx2:
            topic1_words = set(podcast_words.loc[idx1, 'Representation'])
            topic2_words = set(podcast_words.loc[idx2, 'Representation'])
            jaccard_score = len(topic1_words.intersection(topic2_words)) / len(topic1_words.union(topic2_words))
            result.append({'Source1': 'Podcast', 'Topic1': idx1-1, 'Source2': 'Podcast', 'Topic2': idx2-1, 'Similarity': jaccard_score})

    # Internal Comparison within TV Words DataFrame
    for idx1, idx2 in product(tv_words.index, repeat=2):
        if idx1 != idx2:
            topic1_words = set(tv_words.loc[idx1, 'Representation'])
            topic2_words = set(tv_words.loc[idx2, 'Representation'])
            jaccard_score = len(topic1_words.intersection(topic2_words)) / len(topic1_words.union(topic2_words))
            result.append({'Source1': 'TV', 'Topic1': idx1-1, 'Source2': 'TV', 'Topic2': idx2-1, 'Similarity': jaccard_score})

    # Cross-Dataframe Comparison between Email and Podcast Words DataFrames
    for idx1, idx2 in product(email_words.index, podcast_words.index):
        topic1_words = set(email_words.loc[idx1, 'Representation'])
        topic2_words = set(podcast_words.loc[idx2, 'Representation'])
        jaccard_score = len(topic1_words.intersection(topic2_words)) / len(topic1_words.union(topic2_words))
        result.append({'Source1': 'Email', 'Topic1': idx1-1, 'Source2': 'Podcast', 'Topic2': idx2-1, 'Similarity': jaccard_score})

    # Cross-Dataframe Comparison between Email and TV Words DataFrames
    for idx1, idx2 in product(email_words.index, tv_words.index):
        topic1_words = set(email_words.loc[idx1, 'Representation'])
        topic2_words = set(tv_words.loc[idx2, 'Representation'])
        jaccard_score = len(topic1_words.intersection(topic2_words)) / len(topic1_words.union(topic2_words))
        result.append({'Source1': 'Email', 'Topic1': idx1-1, 'Source2': 'TV', 'Topic2': idx2-1, 'Similarity': jaccard_score})

    # Cross-Dataframe Comparison between Podcast and TV Words DataFrames
    for idx1, idx2 in product(podcast_words.index, tv_words.index):
        topic1_words = set(podcast_words.loc[idx1, 'Representation'])
        topic2_words = set(tv_words.loc[idx2, 'Representation'])
        jaccard_score = len(topic1_words.intersection(topic2_words)) / len(topic1_words.union(topic2_words))
        result.append({'Source1': 'Podcast', 'Topic1': idx1-1, 'Source2': 'TV', 'Topic2': idx2-1, 'Similarity': jaccard_score})

    # Create a dataframe from the results
    similarity_df = pd.DataFrame(result)
    return similarity_df
```

```python
similarity_results = calculate_jaccard_similarity(email_words, podcast_words, tv_words)
```

```python
sorted_similarities = similarity_results.sort_values(by='Similarity', ascending=False)
sorted_similarities = sorted_similarities[sorted_similarities['Similarity'] != 0]
sorted_similarities
```

```python
filtered_results = sorted_similarities[(sorted_similarities['Source1'] != sorted_similarities['Source2'])]
```

```python
filtered_results = filtered_results[(filtered_results['Similarity'] >= 0.4) & 
                                    ((filtered_results['Topic1'] != 0) & (filtered_results['Topic2'] != 0))]
```
