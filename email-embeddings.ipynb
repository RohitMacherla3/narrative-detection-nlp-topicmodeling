```python
from sentence_transformers import SentenceTransformer, util
import torch
import pandas as pd
import numpy as np
from concurrent.futures import ThreadPoolExecutor
```

```python
# Define the model

model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
```

```python
def generate_embeddings(text):
    embeddings = model.encode(text, convert_to_tensor=True)
    print("Running...")
    return embeddings
```

```python
emails_data = pd.read_pickle('Email/emails_data_processed.pkl')
emails_data
```

```python
with ThreadPoolExecutor(max_workers=8) as executor:
    embeddings_list = list(executor.map(generate_embeddings, emails_data['Passages']))
```

```python
np.savez('Email/email_embeddings.npz', embeddings_list, allow_pickle=True)
```
