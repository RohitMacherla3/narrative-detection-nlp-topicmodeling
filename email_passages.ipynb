<h2>Import libraries</b>

```python
# First let import the most necessary libs
import pandas as pd
import numpy as np
# Library to import pre-trained model for sentence embeddings
from sentence_transformers import SentenceTransformer
# Calculate similarities between sentences
from sklearn.metrics.pairwise import cosine_similarity
# Visualization library
import seaborn as sns
import matplotlib.pyplot as plt
# package for finding local minimas
from scipy.signal import argrelextrema
import math
```

```python
# Loading a model - don't try it at home, it might take some time - it is 420 mb
model = SentenceTransformer('all-mpnet-base-v2')
```

```python
def concatenate_sentences(sentences):
    
    # Get the length of each sentence
    sentence_lengths = [len(sentence) for sentence in sentences]
    
    # Calculate the mean and standard deviation of sentence lengths
    mean_length = np.mean(sentence_lengths)
    std_dev = np.std(sentence_lengths)
    
    # Determine the longest and shortest outliers
    long_outlier = mean_length + std_dev * 2
    short_outlier = mean_length - std_dev * 2
    
    cleaned_sentences = []
    
    for sentence in sentences:
        if len(sentence) > long_outlier:
            # Replace commas with dots in long sentences
            cleaned_sentence = sentence.replace(',', '.')
        else:
            cleaned_sentence = sentence
        
        cleaned_sentences.append(cleaned_sentence)
    
    # Concatenate the cleaned sentences
    cleaned_text = ' '.join(cleaned_sentences)
    
    return cleaned_text
```

```python
def embed_sentences(text):
    print('Done')
    # Split the text into sentences
    sentences = text.split('. ')
    
    # Embed the sentences using the model
    embeddings = model.encode(sentences)
    
    return embeddings
```

```python
def similarities(embeddings):
    similarities = cosine_similarity(embeddings)
    return similarities
```

<h2>Email Data</h2>

```python
emails_data = pd.read_csv('Email/emails_data_processed.csv')
emails_data
```

```python
emails_data['Body'] = emails_data['Body'].apply(lambda x: x.replace("?", "."))
```

```python
emails_data['Sentences'] = emails_data['Body'].apply(lambda x: x.split('. '))
```

```python
emails_data['cleaned_body'] = emails_data['Sentences'].apply(concatenate_sentences)
```

```python
emails_data['embeddings'] = emails_data['cleaned_body'].apply(embed_sentences)
```

```python
emails_data['similarities'] = emails_data['embeddings'].apply(similarities)
```

```python
def rev_sigmoid(x: float) -> float:
    return (1 / (1 + math.exp(0.5 * x)))

def activate_similarities(similarities: np.array, p_size=1) -> np.array:
    similarities = np.abs(similarities)  # Ensure non-negative similarities
    x = np.linspace(-10, 10, p_size)
    y = np.vectorize(rev_sigmoid)
    activation_weights = np.pad(y(x), (0, similarities.shape[0] - p_size))

    diagonals = [similarities.diagonal(each) for each in range(0, similarities.shape[0])]
    diagonals = [np.pad(each, (0, similarities.shape[0] - len(each))) for each in diagonals]
    diagonals = np.stack(diagonals)

    diagonals = diagonals * activation_weights.reshape(-1, 1)

    activated_similarities = np.sum(diagonals, axis=0)
    
    # Find relative minima using argrelextrema
    minima_indices = argrelextrema(activated_similarities, np.less, order=2)
    
    return minima_indices
```

```python
emails_data['minima'] = emails_data['similarities'].apply(activate_similarities)
```

```python
def passages_split(minmimas, sentences):
    split_points = [each for each in minmimas[0]]
    passages = []
    passage = []
    
    for num, each in enumerate(sentences):
        passage.append(each)
        if num in split_points:
            passages.append(" ".join(passage))
            passage = []
    
    if passage:
        passages.append(" ".join(passage))
    
    return passages
```

```python
emails_data['Passages'] = emails_data.apply(lambda row: passages_split(row['minima'], row['Sentences']), axis=1)
```

```python
emails_data = emails_data.drop(columns=['Body','embeddings', 'similarities', 'minima'])
```
